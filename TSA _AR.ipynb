```python
import yfinance as yf

# Ticker and year
ticker = "RELIANCE.NS"
year = 2005

# Define start and end dates
start_date = f"{year}-01-01"
end_date = f"{year}-12-31"

# Download data
stk_data = yf.download(ticker, start=start_date, end=end_date, progress=False)

# Check if data is available
if stk_data.empty:
    print(f"No data available for {ticker} in {year}.")
else:
    # Keep only OHLC columns
    stk_data = stk_data[["Open", "High", "Low", "Close"]]
    print(f"OHLC data for {ticker} in {year}:")
    print(stk_data)
```

```python
from sklearn.preprocessing import MinMaxScaler
Ms = MinMaxScaler()
data1= Ms.fit_transform(stk_data[[column]])
print("Len:",data1.shape)
```

```python
training_size = round(len(data1 ) * 0.80)
print(training_size)
X_train=data1[:training_size]
X_test=data1[training_size:]
print("X_train length:",X_train.shape)
print("X_test length:",X_test.shape)
y_train=data1[:training_size]
y_test=data1[training_size:]
print("y_train length:",y_train.shape)
print("y_test length:",y_test.shape)
```

```python
pip install --upgrade scikit-learn
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import numpy as np
from statsmodels.tsa.ar_model import AutoReg

# ----- Define Function First -----
def rmsemape(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    print("RMSE-Testset:", rmse)
    
    mape = mean_absolute_percentage_error(y_test, y_pred)
    print("MAPE-Testset:", mape)

# ----- Parameters -----
trends = ['n', 't', 'c', 'ct']
lags = [1, 2, 3, 4]

# ----- Model Loop -----
for i in lags:
    for td in trends:
        
        model = AutoReg(X_train, lags=i, trend=td)
        model_fit = model.fit()
        
        # Make prediction
        y_pred = model_fit.predict(start=len(X_train), end=len(data1)-1)
        
        print("Lag={}, Trend={}".format(i, td))
        rmsemape(y_test, y_pred)
        print("************")
```
